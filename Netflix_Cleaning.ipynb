```python
import pandas as pd
```

```python
# 1. LOAD DATA
df = pd.read_csv('/content/netflix_titles.csv')

# Preview
print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())
```

```python
# Check nulls
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# 2. HANDLE MISSING VALUES


# Fill missing values (example logic)
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)

# Drop rows where critical data is missing
df.dropna(subset=['date_added'], inplace=True)
```

```python
# 3. REMOVE DUPLICATES
duplicates = df.duplicated().sum()
print(f"\nDuplicates found: {duplicates}")

df.drop_duplicates(inplace=True)

# 4. FIX DATA TYPES
# Convert date column to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# 5. CLEAN TEXT DATA
# Standardize text (lowercase + strip spaces)
df['type'] = df['type'].str.lower().str.strip()
df['country'] = df['country'].str.strip()
```

```python
# 6. SPLIT COLUMNS
# Example: split "listed_in" into multiple categories
df['genre_1'] = df['listed_in'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else x)
```

```python
# 7. DROP UNUSED COLUMNS
df.drop(columns=['show_id'], inplace=True)
```

```python
# 8. CREATE NEW FEATURES
# Extract year from date_added
df['year_added'] = df['date_added'].dt.year
```

```python
# 9. FINAL CHECK
print("\nCleaned Data Info:")
print(df.info())

print("\nCleaned Data Sample:")
print(df.head())
```

```python
# 10. SAVE CLEAN DATA
df.to_csv('/content/netflix_titles.csv', index=False)
print("\nCleaned file saved successfully!")
```
